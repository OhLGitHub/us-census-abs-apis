```python


import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

params="NAICS2017,NAICS2017_LABEL,PAYANN,EMP"
location="us:*"
key="d4ceac55dc7bdf83bc5742bf6fe9f5deb05d8b53"
url=f"https://api.census.gov/data/2017/abscs?get={params}&for={location}&key={key}"

full_list=json.loads(requests.get(url).text)
headers = full_list.pop(0)
df = pd.DataFrame(full_list, columns=headers)
df['EMP']=pd.to_numeric(df['EMP'])
df['PAYANN']=pd.to_numeric(df['PAYANN'])
df.drop_duplicates(subset ="NAICS2017_LABEL", keep = False, inplace = True)

df['Average Pay ($1000)'] = (df['PAYANN'] / df['EMP'])
df.drop(['us'],inplace=True,axis=1)
df.drop(['EMP'],inplace=True,axis=1)
df.drop(['PAYANN'],inplace=True,axis=1)
df.drop(['NAICS2017'],inplace=True,axis=1)
df.set_index('NAICS2017_LABEL',inplace=True)
df.drop(labels='Industries not classified', axis=0, inplace=True, errors='raise')
df.drop(labels=None, axis=0, index='Total for all sectors', inplace=True, errors='raise')
df.drop(labels=None, axis=0, index='Other services (except public administration)', inplace=True, errors='raise')
df.dropna(inplace = True)
df.rename(index={'Securities, commodity contracts, and other financial investments and related activities':'Securities and other investments'}, inplace=True)
df.rename(index={'Securities and commodity contracts intermediation and brokerage':'Securities and contract intermediation'}, inplace=True)
df.rename(index={'Research and development in biotechnology (except nanobiotechnology)':'Biotech R&D'}, inplace=True)
df.rename(index={'Sporting goods, hobby, musical instrument, and book stores':'Book and hobby stores'}, inplace=True)
df.sort_values(by='Average Pay ($1000)', inplace=True)
dfTop=df.nlargest(5, 'Average Pay ($1000)')
dfBottom=df.nsmallest(5, 'Average Pay ($1000)')
dfTop.sort_values('Average Pay ($1000)', inplace=True)
dfBottom.sort_values('Average Pay ($1000)', inplace=True)
#df[:]
```

```python
ax3=dfTop.plot(title='Top Five Industries for Average Pay 2017',kind='barh', color='g')
ax3.set_ylabel('Industry')
ax3.set_xlabel('Pay per Employee ($1,000)')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

```python
ax3=dfBottom.plot(title='Bottom Five Industries for Average Pay 2017',kind='barh', color='m')
ax3.set_ylabel('Industry')
ax3.set_xlabel('Pay per Employee ($1,000)')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
